```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
# Step 2: Load the Dataset
file_path = 'C:/Users/HP/Downloads/archive/Tweets.csv' 
tweets_df = pd.read_csv(file_path)
```

```python
# Step 3: Inspect the Data
print("Column names:")
print(tweets_df.columns)
print("\nFirst few rows of the dataset:")
print(tweets_df.head())
```

```python
# Step 4: Data Cleaning and Preparation
# Use relevant columns: 'airline_sentiment' for sentiment and 'text' for tweet text
relevant_columns = ['airline_sentiment', 'text']
tweets_df = tweets_df[relevant_columns].dropna()
```

```python
# Step 5: Sentiment Analysis
# Analyze the sentiment distribution
sentiment_counts = tweets_df['airline_sentiment'].value_counts()
print("\nSentiment distribution:")
print(sentiment_counts)
```

```python
# Visualize sentiment distribution using a bar chart
plt.figure(figsize=(10, 6))
sentiment_counts.plot(kind='bar', color=['red', 'blue', 'green'])
plt.title('Sentiment Distribution')
plt.xlabel('Sentiment')
plt.ylabel('Number of Tweets')
plt.xticks(rotation=0)
plt.show()
```

```python
# Step 6: Visualization - Word Cloud
# Function to generate word cloud
def generate_wordcloud(data, title):
    wordcloud = WordCloud(width=800, height=400, background_color='white').generate(data)
    plt.figure(figsize=(10, 6))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.title(title)
    plt.show()
```

```python
# Generate word clouds for each sentiment category
for sentiment in tweets_df['airline_sentiment'].unique():
    sentiment_text = ' '.join(tweets_df[tweets_df['airline_sentiment'] == sentiment]['text'])
    generate_wordcloud(sentiment_text, f'Word Cloud for {sentiment.capitalize()} Sentiment')

print("Task 4 completed: Analyzing and Visualizing Sentiment Patterns in Social Media Data.")
```
